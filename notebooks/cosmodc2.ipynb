```python
old_mapping = {
    "ra": "ra",
    "dec": "dec",
    "g_ab": "mag_true_g",
    "r_ab": "mag_true_r",
    "i_ab": "mag_true_i",
    "z_ab": "mag_true_z",
    "fluxnorm_bulge": "bulge_to_total_ratio_i",
    "fluxnorm_disk": "bulge_to_total_ratio_i", #TODO: 1 - bulge_to_total_ratio_i
    "fluxnorm_agn": "bulge_to_total_ratio_i", #TODO: set this to zero 
    "pa_disk": "position_angle_true",
    "pa_bulge": "position_angle_true",
    "a_d": "size_disk_true",
    "b_d": "size_minor_disk_true",
    "a_b": "size_bulge_true",
    "b_b": "size_minor_bulge_true",
    "redshift": "redshift",
    "galtileid": "galaxy_id",
    "halo_mass": "halo_mass"
}
```

```python
# columns_needed = [
#     'ra', 'dec', 'halo_mass', 'stellar_mass', 'size_disk_true',
#     'size_minor_true', 'sersic_bulge', 'galaxy_id', 'size_bulge_true',
#     'is_central', 'stellar_mass_disk', 'stellar_mass_bulge', 'sersic_disk', 'halo_id', 'size_true',
#     'redshift_true', 'convergence'
# ]


columns_needed = list(old_mapping.values())
columns_needed.append('halo_mass')
columns_needed.append('halo_id')
columns_needed.append('stellar_mass')
```

```python
dc2 = gc.get_quantities(columns_needed)
```

```python
dc2.keys()
```

```python
dc2_new = {}
for new_key,old_key in old_mapping.items():
    print(new_key,old_key)
    dc2_new[new_key] = dc2[old_key]

dc2_new['fluxnorm_disk'] = 1 - dc2_new['fluxnorm_bulge']
dc2_new['fluxnorm_agn'] = 0 * dc2_new['fluxnorm_bulge']
```

```python
dc2_table = Table(dc2_new)
dc2_table = dc2_table[dc2_table['fluxnorm_bulge'] > 0]
```

```python
dc2_table
```

```python
dc2_table = dc2_table[dc2_table['r_ab'] < 26]
```

```python
rows_number_to_keep = np.random.choice(dc2_table['ra'].size, round(dc2_table['ra'].size/100), replace=False)
sub_sample = dc2_table[rows_number_to_keep]
```

```python
out_path = "../data/dc2_table_subsampled.fits"
sub_sample.write(out_path, format="fits", overwrite=True)
```

```python
out_path = "../data/dc2_table.fits"
dc2_table.write(out_path, format="fits", overwrite=True)
```
