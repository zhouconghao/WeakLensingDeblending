```python
import numpy as np
```

```python
g = np.load('../data/g.npz')['g']
weight = np.load('../data/g.npz')['weights']
```

```python
def run_jackknife(R11, wgts, jackknife):
    print("R11 shape",R11.shape)
    
    n_per = R11.shape[0] // jackknife
    n = n_per * jackknife
    print("jackknife stepsize",n_per)

    R11j = np.zeros(jackknife)
    wgtsj = np.zeros(jackknife)

    loc = 0

    for i in range(jackknife):
        wgtsj[i] = np.sum(wgts[loc:loc + n_per])
        R11j[i] = np.sum(R11[loc:loc + n_per] * wgts[loc:loc + n_per]) / wgtsj[i]
        
        loc += n_per
        
    R11bar = np.sum(R11 * wgts) / np.sum(wgts)
    
    print("R11bar:",R11bar)

    print("R11j:",R11j)

    R11vals = np.zeros(jackknife)

    for i in range(jackknife):
        _wgts = np.delete(wgtsj, i)
        
        R11vals[i] = np.sum(np.delete(R11j,i) * _wgts) / np.sum(_wgts)
    print("R11vals:",R11vals)
    
    jk_variance = np.sqrt((n - n_per) / n * np.sum((R11vals - R11bar)**2))
    print("jk_variance:",jk_variance)
    print("std of the mean:", np.std(R11)/np.sqrt(len(R11)))


run_jackknife(g,weight, 100)
```
